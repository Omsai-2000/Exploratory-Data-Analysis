## **Black Friday Exploratory Data Analysis and Feature Engineering**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Problem Statement

>   A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
# Importing Train dataset
df_train=pd.read_csv('train.csv')

#Importing Test dataset
df_test=pd.read_csv('test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
# Merge both datasets
df=df_train.append(df_test)
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
# User_ID column is not required to build the model. So removing the User_ID column
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# Handling categorical variables Like Age, Gender etc..

df['Gender']=df['Gender'].map({'F':0,'M':1})
```

```python
df.head()
```

```python
df['Age'].unique()
```

```python
# Targeting ordinal encoding
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()

# These 2 categories are sufficient to represent all three categories. So we can add these to existing df.
```

```python
df=pd.concat([df,df_city],axis=1)
```

```python
df.head()
```

```python
# So City_category is not required. So we can drop that column.

df.drop('City_Category',axis=1,inplace=True)
df.head()
```

```python
# Missing values

df.isnull().sum()
```

```python
# Focus on replacing missing values
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
# Replacing missing values with mode (frequently occuring data)

df['Product_Category_2'].mode()[0]
```

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df['Product_Category_3'].mode()[0]
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
# Convert object (Stay_In_Current_City_Years) into integer.
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```

```python
# uint8 --> 8 bit unsigned int converting to int.

df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```

```python
df.info()
```
