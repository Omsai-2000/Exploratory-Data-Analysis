```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(1)
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
#Pie chart- Top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Zomato has huge no.of transaction done in india than USA and UK.

```python
# Numerical variables
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observtions:

1.   When ratings are b/w 4.5 to 4.9 --> Excellent
2.   When ratings are b/w 4 to 4.4 --> very good
3.   When ratings are b/w 3.5 to 3.9 --> Good
4.   When ratings are b/w 2.5 to 3.4 --> Average
5.   When ratings are b/w 0 to 2.4 --> Excellent


> Most of the people given 0 rating.







```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['White','Red','Orange','Yellow','Green','Green'])
plt.title('Zomato Ratings')
```

Observation:

1.   Not rated count is very high
2.   Maximum no.of ratings are b/w 2.5 to 3.4



```python
sns.countplot(x='Rating color',data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
plt.title('Zomato Ratings')
```

```python
final_df.columns
```

```python
# Finding countries that has given 0 ratings
final_df[final_df['Aggregate rating']==0.0].groupby('Country').size().reset_index()
```

Observation:

> Maximum no.of 0 ratings in Indian customers





```python
final_df.info()
```

```python
# Finding which country using what currency
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index().rename(columns={0:'Count'})
```

```python
# Finding Country which has online delivery option
final_df[['Country','Has Online delivery']].groupby(['Country','Has Online delivery']).size().reset_index()
```

Observation:



> Online delivery available in India and UAE



```python
# Creating pie chart for top 5 cities distribution
city_labels=final_df.City.value_counts().index
city_values=final_df.City.value_counts().values
plt.pie(city_values[:5],labels=city_labels[:5],autopct='%1.2f%%')
```

```python
# Finding top 10 Cuisines
Cuisines_labels=final_df. Cuisines.value_counts().index
Cuisines_values=final_df. Cuisines.value_counts().values
plt.pie(Cuisines_values[:10],labels=Cuisines_labels[:10],autopct='%1.2f%%')
```

Observation:



> Most of the food consumers are liking the North Indian Cuisines
